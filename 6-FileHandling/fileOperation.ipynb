### Read the file


#### READ line by line

```python
with open ("example.txt" , "r") as file:
     for line in file:
          print(line.strip()) # strip remove the newline character
```

#### read content

```python
with open ("example.txt" , "r") as file:
    content = file.read()
    print(content)
```

### Writing the file

#### writing lines by overriding

```python
with open("example.txt" , "w") as file:
    file.write("hello World!\n")
```

#### writing lines without overriding

```python
with open("example.txt" , "a") as file:
    file.write("hello World!\n")
```

#### write multiple lines

```python
lines= ["1 line\n","2 line\n" ,"3 line\n" , "4 line\n"]
with open("example.txt" , "a") as file:
    file.writelines(lines)
```

## Reading and writing at the same time
 

#### w+ mode
Creates a new file if the file does not exist.

Overwrites the file if it already exists (old data is deleted).

Allows both writing (write()) and reading (read()) operations.

File pointer starts at the beginning of the file.

```python
with open ("example.txt" , "w+") as file:
    file.write("Hello world!\n")
    file.write("New line\n")
    file.seek(0) # move the cursour to start
    content = file.read()
    print(content)
```
