```python
class dog:
    def __init__(self , name , age):
        self.age = age
        self.name = name
    
d1  = dog("tom" , 4)
d2 = dog("lusi" , 4)
```

methodes

```python
class dog:
    def __init__(self , name , age):
        self.age = age
        self.name = name
    def bark(self):
        print(f"{self.name} says wooof")
    
d1  = dog("tom" , 4)
d2 = dog("lusi" , 4)
```

```python
d1.age
```

```python
d1.bark()
```
