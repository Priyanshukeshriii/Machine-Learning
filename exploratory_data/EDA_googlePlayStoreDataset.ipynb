```python
df['Size'].unique()
```

```python
df['Size'].isnull().sum()
```

```python
df['Size'] = df['Size'].apply(
    lambda x: (
        np.nan if pd.isna(x) or x == 'Varies with device'
        else float(x.replace('M','')) * 1_000 if isinstance(x, str) and 'M' in x
        else x.replace('k','')  if isinstance(x, str) and 'k' in x
        else np.nan
    )
)



```

```python
df['Size'].unique()
```

```python
df['Installs'].unique()
```

```python
df['Price'].unique()
```

```python
char_to_remove =['+',',','$']
columns_to_clean = ['Price', 'Installs']
```

```python
for item in char_to_remove:
    for col in columns_to_clean:
        df[col] = df[col].str.replace(item,'')
```

```python
df['Price'].unique()
```

```python
df['Installs'].unique()
```

```python
df['Installs']=df['Installs'].astype('Int64')
df['Price']= df['Price'].astype('float64')
```

```python
df.info()
```

```python
df['Last Updated'].unique()
```

```python
df['Last Updated']=pd.to_datetime(df['Last Updated'])
df['Day'] = df['Last Updated'].dt.day
df['Month'] = df['Last Updated'].dt.month
df['Year'] = df['Last Updated'].dt.year
```

```python
df[['Day', 'Month', 'Year']]
```

```python
df.info()
```

```python
df.drop('Last Updated', axis= 1,inplace= True)
```

```python
df.to_csv('Google_playStore_cleaned.csv')
```
