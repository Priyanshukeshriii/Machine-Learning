checking if the file exists or not

```python
path = "example1.txt"
if os.path.exists(path):
    print(f"The path '{path}' exists")
else:
    print(f"The path '{path}' doesn't exists")
```

checking the path is a file or dirctory

```python
path = "example.txt"
if os.path.isfile(path):
    print(f"The path {path} is file")
elif os.path.isdir(path):
    print(f"The path {path} is dir")
else:
    print(f"The path {path} is not a file nor dir")

```

Finding the absoulte path using the relative path

```python
rel_path = "example.txt"
abs_path = os.path.abspath(rel_path)
print(abs_path)
```
