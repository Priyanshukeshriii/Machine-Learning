### Exception handling
Exception handling is a mechanism used in programming to handle runtime errors so that the program does not crash and can continue execution normally.

An exception is an error that occurs during the execution of a program, such as dividing by zero, accessing an invalid index, or opening a file that does not exist.

Exception handling allows the programmer to detect errors and take appropriate action using special constructs like try, except, else, and finally.

#### Advantages

Prevents program termination due to errors

Improves program reliability

Helps in debugging and error management

Maintains normal flow of the program

#### Types of Exceptions (Python)

Some common types of exceptions are:

ZeroDivisionError – Occurs when a number is divided by zero.

ValueError – Occurs when a function receives a value of the correct type but an inappropriate value.

TypeError – Occurs when an operation is performed on incompatible data types.

IndexError – Occurs when trying to access an index that is outside the range of a list or tuple.

KeyError – Occurs when a dictionary key is not found.

NameError – Occurs when a variable or name is not defined.

FileNotFoundError – Occurs when a program tries to open a file that does not exist.

ImportError – Occurs when a module cannot be imported.

AttributeError – Occurs when an object does not have the requested attribute.

RuntimeError – Occurs when an error happens during program execution that does not fall into another category.

try , except, else

```python
try  :
    a = int(input("enter a number: "))
    b = 10/a
except ZeroDivisionError as ex: 
    print(ex)
except ValueError as ex:
    print('please enter a number')
    print(ex)
except Exception as ex:
    print("exception occoured")
    print(ex)
else : 
    print(f"The result is {b}") # if some exception occured then else block will not be executed else it will be executed
```

try , except, else ,finally

```python
try  :
    a = int(input("enter a number: "))
    b = 10/a
except ZeroDivisionError as ex: 
    print(ex)
except ValueError as ex:
    print('please enter a number')
    print(ex)
except Exception as ex:
    print("exception occoured")
    print(ex)
else : 
    print(f"The result is {b}") # if some exception occured then else block will not be executed else it will be executed
finally:
    print("execution completed")
```
